```python
sents_map = {
    "negative":0,
    "positive": 2,
    "neutral" : 1
}

tweet_data['sentiment'] = tweet_data['sentiment'].map(sents_map)
```

```python
good_emotions = ['joy', 'anticipation', 'optimism', 'surprise']

def good_bad(column):
    for i, word in column.values:
        for emotion in good_emotions:
            if word == emotion:
                column[i] = 2
            else:
                column[i] = 1
    return column
```

```python
emotion_map = {
    'anger': 1,
    'joy': 2,       
    'anticipation': 2,  
    'disgust': 1,         
    'sadness': 1,
    'optimism': 2,       
    'fear': 1,            
    'surprise': 2 
}

tweet_data['emotion'] = tweet_data['emotion'].map(emotion_map)
```
